# Exploração dos dados

## 1 - Importando as bibliotecas e carregando a base de dados
Nesta etapa vamos carregar as bibliotecas necessárias, importar a base e exibir os primeiros registros dela


```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
df = pd.read_csv("../assets/moons.csv")
df.head()
df.info()
```

## 2 - Visualizar a distribuição dos dados
O Objetivo desta etapa é entender visualmente com os dados estão distribuídos para começar a pensar na clusterização.

```python
# Plotando os dados
plt.scatter(x=df["X"], y=df["Y"])
plt.show()
```

## 3 - Definir e aplciar os métodos de clusterização
A ideia aqui é escolher 2 métodos de clusterização, aplicar eles aos dados e comparar seus resultados

```python
# Definição dos algoritmos
km = KMeans(n_clusters=4, random_state=42)
db = DBSCAN(eps=0.4)
# Aplicar os algoritmos sobre os pontos, obtendo os clusteres resultantes
c_km = km.fit_predict(df)
c_db = db.fit_predict(df)
df["label_kmeans"] = c_km
df["label_dbscan"] = c_db
# Plotando as diferenças
fig = plt.figure( figsize=(12, 8) )
plt.subplot(2, 2, 1)
plt.title("Original")
plt.scatter(x=df["X"], y=df["Y"])
plt.subplot(2, 2, 3)
plt.title("K-means")
plt.scatter(x=df["X"], y=df["Y"], c=df["label_kmeans"])
plt.subplot(2, 2, 4)
plt.title("DBSCAN")
plt.scatter(x=df["X"], y=df["Y"], c=df["label_dbscan"])
plt.show()
```
